# Credit Card Fraud Detection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import gc
from datetime import datetime

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

from catboost import CatBoostClassifier
import xgboost as xgb

sns.set_style("whitegrid")
```

```python
RANDOM_STATE = 42
TEST_SIZE = 0.2
VALID_SIZE = 0.25
NO_JOBS = -1
NUM_ESTIMATORS = 200
RFC_METRIC = "gini"
VERBOSE_EVAL = 50
```

```python
data_df = pd.read_csv("5_Credit_Card_Fraud_Detection.csv")
data_df.head()
```

```python
total = data_df.isnull().sum().sort_values(ascending=False)
percent = (data_df.isnull().sum() / data_df.shape[0] * 100).sort_values(ascending=False)
pd.concat([total, percent], axis=1, keys=["Total", "Percent"]).transpose()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
ax = data_df['Class'].value_counts().plot(kind='bar')
plt.title("Fraud vs Non-Fraud Distribution")

plt.subplot(122)
data_df['Class'].value_counts().plot.pie(autopct="%1.5f%%")
plt.title("Class Share")

plt.show()
```

```python
classes = data_df['Class'].value_counts()
normal_share = classes[0] / len(data_df) * 100
fraud_share = classes[1] / len(data_df) * 100

normal_share, fraud_share
```

```python
data_df['Hour'] = (data_df['Time'] / 3600).astype(int)

hourly = (
    data_df
    .groupby(['Hour', 'Class'])['Amount']
    .agg(['min','max','count','sum','mean','median','var'])
    .reset_index()
)

hourly.columns = ['Hour','Class','Min','Max','Transactions','Sum','Mean','Median','Var']
hourly.head()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18,10))

metrics = ['Sum', 'Transactions', 'Mean', 'Max', 'Median', 'Min']

for ax, metric in zip(axes.flatten(), metrics):
    sns.lineplot(data=hourly[hourly.Class==0], x="Hour", y=metric, ax=ax, label="Normal")
    sns.lineplot(data=hourly[hourly.Class==1], x="Hour", y=metric, ax=ax, label="Fraud", color="red")
    ax.set_title(metric)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x="Class", y="Amount", data=data_df, showfliers=False)
plt.show()

plt.figure(figsize=(14,12))
corr = data_df.corr()
sns.heatmap(corr, cmap="Reds", linewidths=0.1)
plt.title("Feature Correlation")
plt.show()
```

```python
features = data_df.columns
t0 = data_df[data_df.Class == 0]
t1 = data_df[data_df.Class == 1]

fig, axes = plt.subplots(8, 4, figsize=(20,30))
axes = axes.flatten()

for i, col in enumerate(features):
    if t0[col].var() == 0 and t1[col].var() == 0:
        axes[i].text(0.5, 0.5, "Constant Feature", ha="center")
        continue
    sns.kdeplot(t0[col], ax=axes[i], label="Normal")
    sns.kdeplot(t1[col], ax=axes[i], label="Fraud", color="red")
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```

```python
target = "Class"
predictors = [c for c in data_df.columns if c not in ["Class"]]

train_df, test_df = train_test_split(
    data_df, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=data_df[target]
)

train_df, valid_df = train_test_split(
    train_df, test_size=VALID_SIZE, random_state=RANDOM_STATE, stratify=train_df[target]
)
```

```python
rf = RandomForestClassifier(
    n_estimators=NUM_ESTIMATORS,
    random_state=RANDOM_STATE,
    n_jobs=NO_JOBS
)

rf.fit(train_df[predictors], train_df[target])
preds = rf.predict(valid_df[predictors])

roc_auc_score(valid_df[target], preds)
```

```python
ada = AdaBoostClassifier(
    n_estimators=NUM_ESTIMATORS,
    learning_rate=0.8,
    random_state=RANDOM_STATE
)

ada.fit(train_df[predictors], train_df[target])
preds = ada.predict(valid_df[predictors])

roc_auc_score(valid_df[target], preds)
```

```python
cat = CatBoostClassifier(
    iterations=300,
    depth=6,
    learning_rate=0.1,
    eval_metric="AUC",
    random_seed=RANDOM_STATE,
    verbose=50
)

cat.fit(
    train_df[predictors], train_df[target],
    eval_set=(valid_df[predictors], valid_df[target])
)

preds = cat.predict(valid_df[predictors])
roc_auc_score(valid_df[target], preds)
```

```python
dtrain = xgb.DMatrix(train_df[predictors], train_df[target])
dvalid = xgb.DMatrix(valid_df[predictors], valid_df[target])
dtest  = xgb.DMatrix(test_df[predictors], test_df[target])

params = {
    "objective": "binary:logistic",
    "eval_metric": "auc",
    "eta": 0.04,
    "max_depth": 2,
    "subsample": 0.8,
    "colsample_bytree": 0.9,
    "seed": RANDOM_STATE
}

model = xgb.train(
    params,
    dtrain,
    num_boost_round=1000,
    evals=[(dtrain,"train"), (dvalid,"valid")],
    early_stopping_rounds=50,
    verbose_eval=50
)

preds = model.predict(dtest)
roc_auc_score(test_df[target], preds)
```
